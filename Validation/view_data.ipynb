```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open("SHEBA_Data/06-depth_pit.dat") as file:
    arr = file.readlines()
    
thicknesses = np.array([[float(el) for el in line.split(" ") if el] for line in arr])
```

```python
with open("SHEBA_Data/10-temp.dat") as file:
    arr = file.readlines()
    
vals = np.array([[float(el) for el in line.split(" ") if el] for line in arr])
```

```python
with open("SHEBA_Data/12-levels.dat") as file:
    Z = [float(z.replace(' ', '')[:-1]) for z in file.readlines()]
```

```python
len(vals)
```

```python
len(thicknesses)
```

```python
pd.merge(pd.DataFrame(thicknesses), pd.DataFrame(vals), on=0)
```

```python
plt.figure(figsize=(20, 10))
plt.imshow(vals[:, 1:].T, aspect='auto', cmap='rainbow',
           extent=[vals[:, 0][0], vals[:, 0][-1], Z[-1], Z[0]])
plt.plot(thicknesses[:, 0], thicknesses[:, 3], color='white', lw=3, ls=':', label='snow surface')
plt.plot(thicknesses[:, 0], thicknesses[:, 4], color='white', lw=3, ls='--', label='snow-ice interface')
plt.plot(thicknesses[:, 0], thicknesses[:, 5], color='white', lw=3, label='ice base')
plt.xticks(size=15)
plt.yticks(size=15)
plt.legend(loc='lower left', facecolor='black', labelcolor='white', prop={'size': 20})
plt.colorbar()
plt.show()
```
